Question 1: What is Simple Linear Regression?

Ans

Simple Linear Regression is a statistical method used to study the relationship between two variables — one independent variable (X) and one dependent variable (Y). It tries to find a straight line (called the regression line) that best fits the data points and shows how changes in X affect Y.

The equation for simple linear regression is:

Y=a+bX

where:

Y = dependent variable (what we’re trying to predict)

X = independent variable (the predictor)

a = intercept (value of Y when X = 0)

b = slope (shows how much Y changes when X increases by 1 unit)

Question 2: What are the key assumptions of Simple Linear Regression?

Ans
There are a few important assumptions that must be met for Simple Linear Regression to work properly. These help ensure the model gives accurate and reliable results. The key assumptions are:

Linearity:
The relationship between the independent variable (X) and the dependent variable (Y) should be linear — meaning, when X increases, Y changes in a straight-line pattern.

Independence:
The observations (data points) should be independent of each other. One observation should not influence another.

Homoscedasticity:
The variance of the errors (residuals) should be constant across all values of X. In simple words, the spread of errors should be roughly the same throughout the line.

Normality of Errors:
The residuals (differences between actual and predicted values) should be normally distributed.

No or Minimal Multicollinearity:
Since simple linear regression has only one independent variable, this usually isn’t an issue — but in multiple regression, it means the independent variables shouldn’t be highly correlated with each other.

Question 3: What is heteroscedasticity, and why is it important to address in regression
models?

Ans

Heteroscedasticity means that the variance of the errors (residuals) in a regression model is not constant across all levels of the independent variable. In simple words, the spread of the residuals changes — sometimes the errors are small, and other times they get bigger as X increases.

For example, if we are predicting income based on age, the variation in income might be small for younger people but much larger for older people — this shows heteroscedasticity.

It is important to address heteroscedasticity because:

It violates one of the key assumptions of regression (constant variance of errors).

It can make the estimated coefficients less efficient and standard errors unreliable, which means the t-tests and p-values may give wrong conclusions.

It affects the accuracy of predictions and reduces the overall performance of the model.

Question 4: What is Multiple Linear Regression?

Ans

Multiple Linear Regression (MLR) is an extension of simple linear regression that uses two or more independent variables (X₁, X₂, X₃, …) to predict the value of a dependent variable (Y).

The main goal of multiple linear regression is to understand how several factors together influence the outcome and to make better predictions.


Y = dependent variable (the one we want to predict)

X₁, X₂, …, Xn = independent variables (predictors)

a = intercept (value of Y when all X’s are 0)

b₁, b₂, …, bn = coefficients (how much Y changes when each X changes by 1 unit, keeping others constant)

Example:
Predicting a student’s exam score (Y) using their study hours (X₁), attendance percentage (X₂), and sleep hours (X₃).

Question 5: What is polynomial regression, and how does it differ from linear
regression?

ANS

Polynomial Regression is a type of regression that models the relationship between the independent variable (X) and the dependent variable (Y) as an nth degree polynomial. It is used when the data shows a curved (non-linear) relationship rather than a straight line
Difference from Linear Regression:

Linear Regression fits a straight line to the data (relationship between X and Y is linear).

Polynomial Regression fits a curved line by adding higher powers of X, allowing it to capture non-linear patterns in the data.

Example:
If we are predicting the growth of a plant over time, and the growth first increases fast and then slows down, a simple straight line (linear regression) won’t fit well — but a curved line (polynomial regression) will capture that pattern better.

Question 6: Implement a Python program to fit a Simple Linear Regression model to
the following sample data:
● X = [1, 2, 3, 4, 5]
● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
Plot the regression line over the data points.


```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

model = LinearRegression()
model.fit(X, Y)

Y_pred = model.predict(X)

print("Slope (b):", model.coef_[0])
print("Intercept (a):", model.intercept_)

plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.title('Simple Linear Regression')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()
plt.show()
```

Question 7: Fit a Multiple Linear Regression model on this sample data:
● Area = [1200, 1500, 1800, 2000]
● Rooms = [2, 3, 3, 4]
● Price = [250000, 300000, 320000, 370000]


```python

import numpy as np
from sklearn.linear_model import LinearRegression

Area = [1200, 1500, 1800, 2000]
Rooms = [2, 3, 3, 4]
Price = [250000, 300000, 320000, 370000]

X = np.column_stack((Area, Rooms))
Y = np.array(Price)

model = LinearRegression()
model.fit(X, Y)

print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

predicted_price = model.predict([[1600, 3]])  # Example: predict price for 1600 sq.ft and 3 rooms
print("Predicted Price for 1600 sq.ft & 3 rooms:", predicted_price[0])
```

Question 8: Implement polynomial regression on the following data:
● X = [1, 2, 3, 4, 5]
● Y = [2.2, 4.8, 7.5, 11.2, 14.7]
Fit a 2nd-degree polynomial and plot the resulting curve

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
model = LinearRegression()
model.fit(X_poly, Y)
Y_pred = model.predict(X_poly)
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='red', label='Polynomial Curve (Degree 2)')
plt.title('Polynomial Regression (Degree 2)')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()
plt.show()
```

Question 9: Create a residuals plot for a regression model trained on this data:
● X = [10, 20, 30, 40, 50]
● Y = [15, 35, 40, 50, 65]
Assess heteroscedasticity by examining the spread of residuals.


```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])
model = LinearRegression()
model.fit(X, Y)

Y_pred = model.predict(X)

residuals = Y - Y_pred

print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
print("Residuals:", residuals)

plt.scatter(X, residuals, color='blue', label='Residuals')
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Residual Plot')
plt.xlabel('X values')
plt.ylabel('Residuals (Actual - Predicted)')
plt.legend()
plt.show()
```

Question 10: Imagine you are a data scientist working for a real estate company. You
need to predict house prices using features like area, number of rooms, and location.
However, you detect heteroscedasticity and multicollinearity in your regression
model. Explain the steps you would take to address these issues and ensure a robust
model.

Ans

If I were a data scientist working on predicting house prices and I found heteroscedasticity and multicollinearity in my regression model, here’s how I would handle it:

Addressing Heteroscedasticity:

Check residuals: First, I would plot the residuals vs. predicted values to confirm heteroscedasticity (spread of residuals increases or decreases with predictions).

Transform variables: Apply transformations like log, square root, or Box-Cox on the dependent variable (house price) or independent variables to stabilize variance.

Use robust regression: Techniques like Weighted Least Squares (WLS) or robust standard errors can help make the model reliable even if heteroscedasticity exists.

Addressing Multicollinearity:

Check correlations: Compute correlation matrix or Variance Inflation Factor (VIF) to detect highly correlated predictors (like area and number of rooms might be correlated).

Remove or combine variables: If two variables are highly correlated, I could drop one or combine them (e.g., total space instead of area and rooms separately).

Use regularization: Apply Ridge or Lasso regression which penalizes large coefficients and reduces multicollinearity effects.

Other Steps to Ensure a Robust Model:

Feature scaling: Normalize or standardize variables if needed.

Cross-validation: Use techniques like k-fold CV to make sure the model generalizes well.

Outlier detection: Check for extreme values that may distort the model.

Model comparison: Test different models (linear, polynomial, tree-based) to see which gives the most accurate predictions.
