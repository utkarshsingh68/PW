2  How does broadcasting work in NumPy ?
Broadcasting in NumPy is a way of performing arithmetic operations on arrays of different shapes. Normally, NumPy requires arrays to be of the same shape for element-wise operations, but broadcasting automatically “stretches” or “repeats” the smaller array so that it matches the shape of the larger one (without actually copying the data).

Rules of Broadcasting:

NumPy compares shapes from right to left.

If dimensions are equal or one of them is 1, they are compatible.

If they are not compatible, broadcasting will throw an error.

```python
import numpy as np

a = np.array([1, 2, 3])
b = 5

print(a + b)
```

4 Explain the use of the groupby() method in Pandas ?

The groupby() method in Pandas is used to split data into groups based on one or more columns, and then perform operations like sum, mean, count, max, min, etc. on each group.

It follows the Split–Apply–Combine strategy:

Split – divide the DataFrame into groups.

Apply – perform a function (e.g., sum, mean) on each group.

Combine – merge the results back into a new DataFrame.



```python
import pandas as pd

data = {
    "Department": ["IT", "HR", "IT", "Finance", "HR", "Finance"],
    "Employee": ["Aman", "Riya", "Karan", "Meena", "Arjun", "Neha"],
    "Salary": [50000, 40000, 55000, 60000, 42000, 58000]
}

df = pd.DataFrame(data)

result = df.groupby("Department")["Salary"].mean()
print(result)
```

6 A What are the differences between NumPy arrays and Python lists ?
Differences between NumPy Arrays and Python Lists

Data Type

NumPy Array: Stores elements of the same data type (int, float, etc.) → more efficient.

Python List: Can store elements of different data types (int, string, float together).

Performance

NumPy Array: Much faster because it is implemented in C and supports vectorized operations.

Python List: Slower because operations happen element by element using Python loops.

Memory Usage

NumPy Array: Uses less memory (compact storage).

Python List: Uses more memory (stores extra information for each element).

Operations

NumPy Array: Supports element-wise mathematical operations directly (like a+b, a*2).

Python List: Does not support element-wise operations (requires loops).

Functionality

NumPy Array: Comes with built-in functions for linear algebra, statistics, broadcasting, etc.

Python List: Limited functionality (mainly general-purpose storage).

10 What is the significance of hierarchical indexing in Pandas ?
Significance of Hierarchical Indexing in Pandas

Hierarchical indexing (MultiIndexing) in Pandas means having multiple levels of index (row or column labels) in a DataFrame or Series.

It allows us to work with higher-dimensional data (2D, 3D) in a compact 2D table format.

Why it is significant:

Organizes complex data – Makes it easier to represent data with multiple keys (e.g., Country → State → City).

Easier data analysis – Enables advanced grouping, slicing, and reshaping of data.

Supports pivoting – Useful in pivot tables and when working with multi-level groupby results.

Compact representation – Avoids the need for multiple DataFrames to represent multi-dimensional relationships.


```python
import pandas as pd

data = {
    "Sales": [250, 300, 400, 350],
    "Profit": [50, 70, 100, 90]
}

index = pd.MultiIndex.from_tuples(
    [("USA", "New York"), ("USA", "Texas"), ("India", "Delhi"), ("India", "Mumbai")],
    names=["Country", "City"]
)

df = pd.DataFrame(data, index=index)
print(df)
```

13  Why is handling missing data important in Pandas ?
In real-world datasets, missing values are very common (due to human error, data collection issues, or system problems). Handling missing data is important because:

Accuracy of Analysis – Missing values can lead to wrong averages, correlations, or predictions if not treated properly.

Model Training – Many machine learning algorithms cannot work with missing data, so it must be cleaned before training.

Consistency – Missing values may cause errors in calculations (like NaN in sums, means, etc.).

Bias Reduction – Ignoring missing values incorrectly can introduce bias into results.

Data Integrity – Cleaning missing data ensures the dataset is reliable and complete for decision-making.

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Aman", "Riya", "Karan", "Meena"],
    "Marks": [85, np.nan, 90, np.nan]
}

df = pd.DataFrame(data)

print(df.isnull())
print(df.fillna(0))
print(df.dropna())
```

19 How does Pandas simplify time series analysis ?

Pandas provides special features and functions for working with time-based data, making time series analysis much easier compared to plain Python.

Key Ways Pandas Helps:

Datetime Handling – Pandas has a powerful DatetimeIndex and to_datetime() function to easily convert strings into proper datetime objects.

Resampling – Allows changing the frequency of data (e.g., daily to monthly, hourly to weekly) using .resample().

Shifting and Lagging – Provides .shift() to move data backward or forward in time for trend and lag analysis.

Rolling & Window Functions – Supports moving averages, rolling sums, and statistical calculations with .rolling().

Time-Based Indexing & Slicing – Enables selecting data by specific dates, ranges, or even years/months.

Built-in Plotting – Easy visualization of time series using .plot().




```python
import pandas as pd

dates = pd.date_range("2023-01-01", periods=6, freq="D")
data = pd.Series([10, 15, 20, 25, 30, 35], index=dates)

print(data.resample("2D").mean())  # Resampling to 2-day frequency
print(data.shift(1))               # Shifting data by 1 day
print(data.rolling(3).mean())      # 3-day moving average
```

21 Why is NumPy’s array slicing faster than Python’s list slicing ?

NumPy arrays are much faster than Python lists when it comes to slicing operations because of the way they are stored and handled in memory.

Main Reasons:

Contiguous Memory Storage –
NumPy arrays are stored in a single continuous block of memory, while Python lists are stored as arrays of pointers to objects scattered in memory. This makes accessing and slicing NumPy arrays much faster.

Homogeneous Data Type –
NumPy arrays only store elements of the same type (e.g., all integers or all floats). This allows efficient memory usage and optimized operations. In contrast, Python lists can hold mixed types, making slicing slower.

Views Instead of Copies –
In NumPy, slicing generally returns a view of the original array (not a full copy). This avoids extra memory usage and speeds up execution. Python list slicing, however, creates a new list, which takes more time and memory.

Low-level Optimizations (C/Fortran) –
NumPy is implemented in C and uses optimized algorithms at a low level. Python lists are higher-level objects managed by Python’s interpreter, which adds overhead.



```python
import numpy as np
import time

# NumPy array slicing
arr = np.arange(1_000_000)
start = time.time()
slice_np = arr[100:200000]
end = time.time()
print("NumPy slicing time:", end - start)

# Python list slicing
lst = list(range(1_000_000))
start = time.time()
slice_lst = lst[100:200000]
end = time.time()
print("List slicing time:", end - start)
```
