In Python, when we create a function, we use parameters as placeholders for the values the function will receive. These are written inside the brackets when we define the function.

When we call the function, we pass arguments, which are the actual values that get assigned to the parameters.

So, parameters are used while defining the function, and arguments are used while calling the function.

```python
def multiply(x, y):
    return x * y

result = multiply(4, 6)
print(result)
```

3. What are the different ways to define and call a function in Python?

In Python, we can define and call functions in different ways depending on how we pass arguments. Here are the main ways:

```python
# 1. Normal function with positional argument
def greet(name):
    print("Hello", name)

# 2. Function with default argument
def greet_default(name="Friend"):
    print("Hello", name)

# 3. Function with keyword arguments
def info(name, age):
    print(name, "is", age, "years old")

# 4. Function with variable-length arguments (*args)
def total(*numbers):
    print("Total is", sum(numbers))

# 5. Function with keyword variable-length arguments (**kwargs)
def details(**data):
    print("Details:", data)

# 6. Lambda function (anonymous function)
square = lambda x: x * x

# Function Calls
greet("Utkaarsh")
greet_default()
greet_default("Anjali")
info(age=22, name="Utkarsh")
details(name="Utkarsh", age=22, city="Bhopal")
print("Square of 5 is", square(5))
```

6. Explain the concept of generators in Python and how they are defined?

A generator is a special type of function in Python that gives one value at a time, using the yield keyword instead of return. It remembers where it left off and continues from there when called again.

Generators are memory-efficient because they don’t store all values at once—they generate them on the fly.

```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

numbers = count_up_to(3)

print(next(numbers))
print(next(numbers))
print(next(numbers))
```

7. What are the advantages of using generators over regular functions?


Generators are better than regular functions when we want to save memory and get values one by one. They don’t create the whole list in memory—they generate values on the fly using yield.

Advantages of Generators:
Memory Efficient – They don’t store all values in memory.

Faster for large data – Values are calculated only when needed.

Pause and Resume – They remember where they left off using yield.

Cleaner Code – Useful when looping through big data or streams.

```python
# Regular function returns full list
def get_list(n):
    result = []
    for i in range(n):
        result.append(i)
    return result

# Generator returns one value at a time
def get_gen(n):
    for i in range(n):
        yield i

# Using generator
gen = get_gen(5)
print(next(gen))
print(next(gen))
```

8. What is a lambda function in Python and when is it typically used?

A lambda function is a small, one-line anonymous function in Python. It doesn't have a name like normal functions made with def, and it’s used when we need a quick function for a short task.

We use lambda functions mostly when:

The function is very simple

We need it only once or temporarily

We use it with functions like map(), filter(), or sorted()

```python
nums = [1, 2, 3, 4, 5, 6]
even = list(filter(lambda x: x % 2 == 0, nums))
print(even)
```

9. Explain the purpose and usage of the `map()` function in Python.?

The map() function in Python is used to apply a function to every item in a list (or any iterable). It returns a new map object with the results, which we usually convert to a list.

It’s useful when we want to do the same operation to each item in a list — like doubling numbers, converting to uppercase, etc.

```python
nums = [1, 2, 3, 4]
doubled = list(map(lambda x: x * 2, nums))
print(doubled)
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

```python
#1. map() – Changes each item
nums = [1, 2, 3, 4]
squared = list(map(lambda x: x * x, nums))
print(squared)  # [1, 4, 9, 16]
```

```python
# 2. filter() – Keeps only matching items
nums = [1, 2, 3, 4, 5, 6]
even = list(filter(lambda x: x % 2 == 0, nums))
print(even)  # [2, 4, 6]
```

```python
#reduce() – Combines all items into one
from functools import reduce

nums = [1, 2, 3, 4]
total = reduce(lambda x, y: x + y, nums)
print(total)  # 10
```

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_of_even(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total


my_list = [10, 15, 22, 33, 40]
result = sum_of_even(my_list)
print("Sum of even numbers:", result)
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

for num in range(1, 201):
    if is_prime(num):
        print(num, "is a prime number")
    else:
        print(num, "is not a prime number")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.a = 0
        self.b = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            next_val = self.a + self.b
            self.a = self.b
            self.b = next_val
            self.count += 1
            return next_val


fib = Fibonacci(10)
for num in fib:
    print(num)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent

```python
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

for value in powers_of_two(5):
    print(value)
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()

# Example usage
for line in read_lines("sample.txt"):
    print(line)
```

```python
celsius = [0, 10, 20, 30, 40]

fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print("Temperatures in Fahrenheit:", fahrenheit)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(text):
    vowels = "aeiouAEIOU"
    result = ''.join(filter(lambda ch: ch not in vowels, text))
    return result

input_str = "Utkarsh Kumar"
output_str = remove_vowels(input_str)
print("String without vowels:", output_str)
```
