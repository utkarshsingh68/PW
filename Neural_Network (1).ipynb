Question 1: What is Deep Learning? Briefly describe how it evolved and how it differs
from traditional machine learning.

ANS

Deep Learning is a subset of Machine Learning that uses Artificial Neural Networks with many layers (called deep neural networks) to learn patterns from data automatically. It is inspired by the structure and working of the human brain. Deep learning models can learn complex patterns from large amounts of data such as images, speech, and text.

Early Stage (1940s–1980s):
Scientists introduced the concept of artificial neurons, like the Perceptron. However, these models were simple and had limited capability.

Machine Learning Growth (1990s–2000s):
Traditional machine learning algorithms like Decision Trees, SVM, and Logistic Regression became popular. Neural networks existed but were not widely used due to low computing power and lack of large datasets.

Revival of Neural Networks (2006):
Geoffrey Hinton and other researchers introduced techniques to train deep neural networks effectively. This marked the beginning of modern deep learning.

Breakthrough Period (2012–Present):
With the availability of large datasets, powerful GPUs, and improved algorithms, deep learning achieved great success in image recognition, speech recognition, and natural language processing. Models like CNN, RNN, and Transformers became widely used.

Difference between Traditional Machine Learning and Deep Learning

Feature Extraction:

Traditional Machine Learning: Features are extracted manually by humans.

Deep Learning: Features are extracted automatically by the model.

Data Requirement:

Traditional Machine Learning: Works well with small or medium datasets.

Deep Learning: Requires large amounts of data.

Model Complexity:

Traditional Machine Learning: Uses simple models like Decision Trees, SVM, Linear Regression.

Deep Learning: Uses complex neural networks with multiple layers.

Performance on Complex Tasks:

Traditional Machine Learning: Less effective for complex tasks like image and speech recognition.

Deep Learning: Very effective for complex tasks.

Hardware Requirement:

Traditional Machine Learning: Can run on normal CPUs.

Deep Learning: Requires powerful hardware like GPUs.

Training Time:

Traditional Machine Learning: Takes less time to train.

Deep Learning: Takes more time to train.

Automation Level:

Traditional Machine Learning: Requires more human intervention.

Deep Learning: Requires less human intervention.

Examples:

Traditional Machine Learning: Spam detection, price prediction.

Deep Learning: Image recognition, voice assistants, self-driving cars.

Question 2: Explain the basic architecture and functioning of a Perceptron. What are its
limitations?

Ans

Basic Architecture of a Perceptron :

The Perceptron is the simplest type of artificial neural network.

It consists of the following main components:

Input Layer: Receives input values (x₁, x₂, x₃, …, xₙ).

Weights (w₁, w₂, w₃, …, wₙ): Each input has a weight that shows its importance.

Bias (b): A constant value added to adjust the output.

Summation Unit: Multiplies inputs with weights and adds bias.

Activation Function: Produces the final output (usually Step Function).



Functioning of a Perceptron :

The perceptron receives multiple input values.

Each input is multiplied by its corresponding weight.

All weighted inputs are added together with the bias.

The result is passed through an activation function.

The activation function produces the final output:

Output = 1 (if value ≥ threshold)

Output = 0 (if value < threshold)

It is mainly used for binary classification (Yes/No, True/False).


Limitations of Perceptron :

It can only solve linearly separable problems.

It cannot solve non-linear problems like XOR.

It uses only a single layer, so it cannot learn complex patterns.

It has limited learning capability.

It produces only binary output (0 or 1).

It is not suitable for complex real-world problems like image recognition.

Question 3: Describe the purpose of activation function in neural networks. Compare
Sigmoid, ReLU, and Tanh functions.

Ans

The activation function in neural networks is used to determine whether a neuron should be activated and to introduce non-linearity into the model. Without an activation function, the neural network would behave like a simple linear model and would not be able to learn complex patterns. Activation functions transform the input signal into an output signal that can be passed to the next layer, allowing the network to solve complex problems such as image recognition and speech processing. They also help improve the learning capability and overall performance of the neural network.

The Sigmoid, Tanh, and ReLU are commonly used activation functions. The Sigmoid function produces output values between 0 and 1, making it useful for binary classification problems, but it suffers from the vanishing gradient problem, which slows down learning. The Tanh function produces output values between −1 and +1 and is better than Sigmoid because it is zero-centered, which helps in faster and more stable learning, but it still has the vanishing gradient problem. The ReLU (Rectified Linear Unit) function produces output values from 0 to infinity and is the most widely used activation function in deep learning because it is simple, fast, and reduces the vanishing gradient problem. However, it can sometimes cause the dead neuron problem. Overall, ReLU is preferred in most deep learning models due to its efficiency and better performance.


Question 4: What is the difference between Loss function and Cost function in neural
networks? Provide examples.

Ans

In neural networks, both the loss function and cost function are used to measure how well the model is performing, but they differ in their scope. The **loss function** measures the error for a single training example. It calculates the difference between the predicted output and the actual output for one data point. For example, in a house price prediction, the loss function measures the error between the predicted price and the actual price of one house. Examples of loss functions include **Mean Squared Error (MSE)** for regression and **Binary Cross-Entropy** for binary classification.

On the other hand, the **cost function** measures the average error over the entire training dataset. It is the overall error of the model and is calculated by taking the average of all individual loss values. The cost function helps in evaluating the overall performance of the neural network and is used during training to update the model weights using optimization algorithms like Gradient Descent. For example, if there are 100 training examples, the cost function is the average of the loss values of all 100 examples. An example is the **Average Mean Squared Error**, which calculates the average squared error across all training samples.

In summary, the loss function calculates the error for a single example, while the cost function calculates the average error for the entire dataset. Both are important for training neural networks, as the loss function measures individual errors and the cost function helps optimize the overall model performance.


Question 5: What is the role of optimizers in neural networks? Compare Gradient
Descent, Adam, and RMSprop.

Ans

Optimizers play a crucial role in neural networks by helping the model learn the best weights and biases that minimize the error. The main purpose of an optimizer is to reduce the loss function by adjusting the network parameters during training. It works by calculating the gradient (slope) of the loss function and updating the weights in the direction that reduces the error. This process is repeated many times until the model reaches the minimum error and gives accurate predictions. Without an optimizer, the neural network cannot learn or improve its performance.

Gradient Descent is the basic and most commonly used optimizer. It updates the weights by moving in the direction of the negative gradient of the loss function. It is simple and easy to understand but can be slow and may get stuck in local minima. RMSprop (Root Mean Square Propagation) is an improved optimizer that adjusts the learning rate automatically for each parameter. It helps in faster and more stable training compared to basic Gradient Descent and works well for complex and deep neural networks. Adam (Adaptive Moment Estimation) is one of the most advanced and widely used optimizers. It combines the advantages of both Gradient Descent and RMSprop by using adaptive learning rates and momentum. Adam provides faster convergence, better performance, and is highly efficient for most deep learning applications. Therefore, while Gradient Descent is simple, RMSprop improves learning efficiency, and Adam is generally the best choice due to its speed and accuracy.

Question 6: Write a Python program to implement a single-layer perceptron from
scratch using NumPy to solve the logical AND gate.


```python
import numpy as np

X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 0, 0, 1])

weights = np.zeros(2)
bias = 0
learning_rate = 0.1
epochs = 10

def step_function(x):
    return 1 if x >= 0 else 0

for epoch in range(epochs):
    for i in range(len(X)):
        weighted_sum = np.dot(X[i], weights) + bias
        y_pred = step_function(weighted_sum)
        error = y[i] - y_pred
        weights = weights + learning_rate * error * X[i]
        bias = bias + learning_rate * error

for i in range(len(X)):
    weighted_sum = np.dot(X[i], weights) + bias
    y_pred = step_function(weighted_sum)
    print(X[i], "→", y_pred)
```

Question 7: Implement and visualize Sigmoid, ReLU, and Tanh activation functions
using Matplotlib.


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

sigmoid = 1 / (1 + np.exp(-x))
relu = np.maximum(0, x)
tanh = np.tanh(x)

plt.figure(figsize=(10, 6))

plt.plot(x, sigmoid, label="Sigmoid")
plt.plot(x, relu, label="ReLU")
plt.plot(x, tanh, label="Tanh")

plt.title("Activation Functions")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.grid()

plt.show()
```

Question 8: Use Keras to build and train a simple multilayer neural network on the
MNIST digits dataset. Print the training accuracy

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0

model = keras.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(X_train, y_train, epochs=5, batch_size=32)

print("Training Accuracy:", history.history['accuracy'][-1])
```

Question 9: Visualize the loss and accuracy curves for a neural network model trained
on the Fashion MNIST dataset. Interpret the training behavior

Ans

The loss and accuracy curves help us understand how well the neural network is learning. As training progresses, the training loss decreases and the training accuracy increases, which indicates that the model is learning from the data. Similarly, the validation loss should decrease and validation accuracy should increase if the model is generalizing well. If the training accuracy becomes very high but validation accuracy remains low, it indicates overfitting. If both accuracies are low, it indicates underfitting. In a well-trained model, both training and validation accuracy increase steadily and the loss decreases, showing good learning and generalization performance.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt

(X_train, y_train), (X_test, y_test) = keras.datasets.fashion_mnist.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0

model = keras.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Curve')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy Curve')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```

Question 10: You are working on a project for a bank that wants to automatically detect
fraudulent transactions. The dataset is large, imbalanced, and contains structured
features like transaction amount, merchant ID, and customer location. The goal is to
classify each transaction as fraudulent or legitimate.
Explain your real-time data science workflow:
● How would you design a deep learning model (perceptron or multilayer NN)?
● Which activation function and loss function would you use, and why?
● How would you train and evaluate the model, considering class imbalance?
● Which optimizer would be suitable, and how would you prevent overfitting?


Ans

1. Designing the Deep Learning Model

For fraud detection, I would use a Multilayer Neural Network (MLP) because the dataset is large and contains complex structured features like transaction amount, merchant ID, and customer location. A single perceptron is too simple and cannot learn complex patterns. The model would have an input layer that takes all transaction features, followed by multiple hidden layers with several neurons to learn complex relationships between features, and an output layer with one neuron to classify the transaction as fraudulent (1) or legitimate (0). The hidden layers help the model learn important patterns such as unusual transaction behavior.

2. Activation Function and Loss Function

I would use the ReLU (Rectified Linear Unit) activation function in the hidden layers because it is fast, efficient, and helps in learning complex patterns without the vanishing gradient problem. In the output layer, I would use the Sigmoid activation function because this is a binary classification problem, and Sigmoid gives output between 0 and 1, which can be interpreted as probability.

For the loss function, I would use Binary Cross-Entropy Loss because it is the most suitable loss function for binary classification problems. It measures the difference between the predicted probability and the actual class label, helping the model learn accurately.

3. Training and Evaluation Considering Class Imbalance

Since the dataset is imbalanced (fraud cases are rare compared to legitimate transactions), I would use techniques such as class weighting, where higher weight is given to fraud cases so the model pays more attention to them. I would also use techniques like oversampling (SMOTE) or undersampling to balance the dataset. For evaluation, I would not rely only on accuracy because accuracy can be misleading in imbalanced datasets. Instead, I would use Precision, Recall, F1-score, and ROC-AUC score, which give a better understanding of fraud detection performance, especially recall, because detecting fraud is very important.

4. Optimizer and Preventing Overfitting

I would use the Adam optimizer because it is fast, efficient, and works well for most deep learning problems. It automatically adjusts the learning rate and improves convergence speed.

To prevent overfitting, I would use techniques such as:

Dropout, which randomly disables some neurons during training

Early stopping, which stops training when validation performance stops improving

Regularization (L1/L2) to reduce model complexity

Using validation data to monitor model performance
